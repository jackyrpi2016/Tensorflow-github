```python
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = np.random.randint(0, 100, (10, 2))
```

```python
data
```

scale all the data

scale_model is an instance of MinMaxScaler()

```python
scaler_model = MinMaxScaler()
```

```python
type(scaler_model)
```

!!! no need to care about warning

just convert int32 to float64

```python
scaler_model.fit(data)
```

transform the data: divide all the data by the max value (98)

```python
scaler_model.transform(data)
```

always, we fit the training data, but we transform both the training and the test data.

If you want to fit and transform the training at the same time. use fit_transform()

```python
scaler_model.fit_transform(data)
```

```python
import pandas as pd
```

```python
mydata = np.random.randint(0, 100, (50, 4))
```

```python
mydata
```

```python
df = pd.DataFrame(data = mydata)
```

```python
df
```

```python
df = pd.DataFrame(data=mydata, columns = ['f1', 'f2', 'f3', 'label'])
```

```python
df
```

```python
X = df[['f1', 'f2', 'f3']]
```

```python
y = df['label']
```

```python
from sklearn.model_selection import train_test_split
```

train : test = 7 : 3, random_state is just random seed

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```
